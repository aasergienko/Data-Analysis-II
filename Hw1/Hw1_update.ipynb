```python
from pymystem3 import Mystem
import pymorphy2
with open("text.txt", encoding="utf-8") as f:
    text = str(f.read())
a = text.split("\n")
text = " ".join(a)

#здесь я читаю текст
```

```python
m = Mystem()
lemmas = m.lemmatize(text)
#произошла лемматизация
```

```python
new_lemmas = []
for word in lemmas:
    fl = False
    for char in word:
        if (ord(char) >= ord('a') and ord(char) <= ord("я")) or (ord(char) >= ord('А') and ord(char) <= ord("Я")):
            fl = True
    if fl:
        new_lemmas.append(word)
        
#я чищу список от знаков препинания
```

```python
with open ('lemmas.txt', 'w') as f:
    f.write("\n".join(new_lemmas))
```

```python
with open ('lemmas.txt', 'r') as f:
    text = f.read()
    lemmas_2 = text.splitlines()
#читаю файл
```

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
analyzed = []
for word in lemmas_2:
    for p in morph.parse(word):
        dct = {
            "lemma": p.normal_form,
            "word": p.word,
            "pos": p.tag.POS
           }
    analyzed.append(dct)
print (analyzed[30:40])

#здесь я создаю список из словарей с ключами "lemma", 'word', 'pos'
```

```python
import json
with open('parsed.json', 'w') as f:
    json.dump(analyzed, f)
#сохраняю в json
```

```python
import collections
POSspisok = []
for word in analyzed:
    POSspisok.append(word['pos'])
    
counter = collections.Counter(POSspisok)
print (counter)

#считать проценты каждой части речи было лень, простите
    
```

```python
from collections import Counter
c = Counter()
for d in analyzed:
    if d['pos'] == 'INFN':
        verbs = d['word']
        c[verbs] += 1
print(c.most_common(20))

#самые частотные глаголы
```

```python
from collections import Counter
c = Counter()
for d in analyzed:
    if d['pos'] == 'ADVB':
        advb = d['word']
        c[advb] += 1
print(c.most_common(20))

#самые частотные наречия
```

```python
import nltk
b = Counter()
bigrm = list(nltk.bigrams(lemmas_2))
for d in bigrm:
    bigram = d
    b[bigram] +=1
print(b.most_common(25))

#самые частотные биграммы
```
